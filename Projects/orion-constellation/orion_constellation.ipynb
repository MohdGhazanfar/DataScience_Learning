The goal of the project is to understand spatial perspective. 

We visualize Orion in both 2D and 3D, and will be able to see the difference in the constellation shape humans see from earth versus the actual position of the stars that make up this constellation.

## 2. Familiarize with real data

Astronomers describe a star's position in the sky by using a pair of angles: declination and right ascension. Declination is similar to longitude, but it is projected on the celestian fear. Right ascension is known as the "hour angle" because it accounts for time of day and earth's rotaiton. Both angles are relative to the celestial equator. You can learn more about star position [here](https://en.wikipedia.org/wiki/Star_position).

The `x`, `y`, and `z` lists below are composed of the x, y, z coordinates for each star in the collection of stars that make up the Orion constellation as documented in a paper by Nottingham Trent Univesity on "The Orion constellation as an installation" found [here](https://arxiv.org/ftp/arxiv/papers/1110/1110.3469.pdf).

````python
# Orion
x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25,13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25,1.38]
````

## 3. Creating a 2D Visualization

Before we visualize the stars in 3D, let's get a sense of what they look like in 2D. 

We create a figure for the 2d plot and save it to a variable name `fig`. 

We add the subplot `.add_subplot()` as the single subplot, with `1,1,1`.

With the scatter [function](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html) to visualize your `x` and `y` coordinates. (using: `.scatter(x,y)`)

Render the visualization.

Below we use a 2D visualization of the Orion constellation, but it looks very different to that we would see in the night sky and unrecognizable. There is a curve to the sky, and this is a flat visualization, but we will visualize it in 3D in the next step to get a better sense of the actual star positions. 

````python
fig = plt.figure()
fig.add_subplot(1, 1, 1)
plt.scatter(x, y)
plt.show()
````

## 4. Creating a 3D Visualization

We create a figure for the 3D plot and save it to a variable name `fig_3d`.

Since this will be a 3D projection, we want to make to tell Matplotlib this will be a 3D plot.  

To add a 3D projection, we must include a the projection argument. It would look like this:
```py
projection="3d"
```

Add the subplot with `.add_subplot()` as the single subplot `1,1,1` and we specify the `projection` as `3d`:

`fig_3d.add_subplot(1,1,1,projection="3d")`)

Since this visualization will be in 3D, we will need our third dimension. In this case, our `z` coordinate. 

We create a new variable `constellation3d` and call the scatter [function](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html) with your `x`, `y` and `z` coordinates. 

Then we include `z` just as we have been including the other two axes. (hint: `.scatter(x,y,z)`)

Render the visualization.


````python
fig_3d = plt.figure()
fig_3d.add_subplot(1, 1, 1, projection="3d")
constellation = plt.scatter(x, y, z)
plt.show()
plt.title('Orion Constellation 3D Scatter Plot')
````
