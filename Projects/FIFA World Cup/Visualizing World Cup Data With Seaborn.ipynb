```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('WorldCupMatches.csv')
print(df.head())
```

```python
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
print(df['Total Goals'])
```

```python
# styling
sns.set_style("whitegrid")
sns.set_context('poster', font_scale=0.8)
```

```python
# subplot axes and figure
f, ax = plt.subplots(figsize=(20, 20))
ax = sns.barplot(x=df['Year'], y=df['Total Goals'])
plt.show()
ax.set_title('Avg. Goals per year in World Cup')
```

```python
df_goals = pd.read_csv('goals.csv')
print(df_goals.head())
```

```python
sns.set_context('notebook', font_scale=1.25)
```

```python
f, ax2=plt.subplots(figsize=(15, 15))
ax2=sns.boxplot(x='year', y='goals', data=df_goals)
plt.show()
ax2.set_title('Goals Visualisation')
```
